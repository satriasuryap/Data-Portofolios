# Data Cleaning and Pre-Processing

Just to be safe it is recomended to check for duplicates inside the dataframe with drop_duplicates() functions

```python
data.drop_duplicates()
```

Since we can see with our eyes that a column has a lot of Null values we will check whether it is fully null or not.

```python
nan_count = data.isnull().sum()
print(nan_count)
```

And as we can see most of the data in the *RETURNTOPLAY* column has mostly null values it is safe for us to drop the column since it will not be affect any harm on the dataframe, and we will also drop the *LEAGUE_ID* column since every game on the NBA is played on the same league so it is safe to drop the column as well.

```python
columns = ['LEAGUE_ID', 'RETURNTOPLAY']
dropped_data = data.drop(columns, axis = 1)
dropped_data
```

On the *HOME_RECORD* and *ROAD_RECORD* column it is clearly stated on how much does the team win or lose with each home and away games, but won't it be easier if the two can be separated and maybe can be build a query in the future?

```python
dropped_data[['HOME_WIN', 'HOME_LOSE']] = dropped_data['HOME_RECORD'].str.split('-', expand=True)
```

```python
dropped_data[['ROAD_WIN', 'ROAD_LOSE']] = dropped_data['ROAD_RECORD'].str.split('-', expand=True)
```

```python
dropped_data
```

On the outsiders perspective it will be really hard to look at the data and point out on which seasons are these data? since a single season of the NBA goes from late October to the middle of April, so I think it will be easier if people can see which seasons are these rankings belong to.

```python
conditions = [
    (dropped_data['SEASON_ID'] == 22013),(dropped_data['SEASON_ID'] == 22014),(dropped_data['SEASON_ID'] == 22015),
    (dropped_data['SEASON_ID'] == 22016),(dropped_data['SEASON_ID'] == 22017),(dropped_data['SEASON_ID'] == 22018),
    (dropped_data['SEASON_ID'] == 22019),(dropped_data['SEASON_ID'] == 22020),(dropped_data['SEASON_ID'] == 22021),
    (dropped_data['SEASON_ID'] == 22022)
    ]

values = ['13-14', '14-15', '15-16', '16-17', '17-18', '18-19', '19-20', '20-21', '21-22', '22-23']

dropped_data['SEASONS'] = np.select(conditions, values)

dropped_data
```
